```python
docs = [
    "The college library is open from 9 AM to 8 PM on weekdays.",
    "Students must carry their ID card to enter the campus.",
    "Mid-semester exams start from October 15.",
    "The cafeteria closes at 9 PM."
]
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=20
)

documents = text_splitter.create_documents(docs)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vector_db = FAISS.from_documents(documents, embeddings)
```

```python
from transformers import pipeline

generator = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=150
)
```

```python
query = "What time does the library open?"

retrieved_docs = vector_db.similarity_search(query, k=2)

context = " ".join([doc.page_content for doc in retrieved_docs])

prompt = (
    f"Answer the question using the context below.\n\n"
    f"Context: {context}\n\n"
    f"Question: {query}\n"
    f"Answer:"
)

response = generator(prompt)
print(response[0]["generated_text"])
```

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("COC.pdf")
documents = loader.load()
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
from langchain_community.vectorstores import FAISS

vector_db = FAISS.from_documents(chunks, embeddings)
```

```python
from transformers import pipeline

generator = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=150
)
```

```python
query = "What is the maximum monetary penalty fine that can be imposed on a student?"

retrieved_docs = vector_db.similarity_search(query, k=2)
context = " ".join(doc.page_content for doc in retrieved_docs)

prompt = (
    f"Answer the question using the context below.\n\n"
    f"Context: {context}\n\n"
    f"Question: {query}\n"
    f"Answer:"
)

response = generator(prompt)
print(response[0]["generated_text"])
```

```python
##LEVEL3: Agentic Campus Assisstant
```

```python
def search_pdf(query):
    docs = vector_db.similarity_search(query, k=2)
    return " ".join(doc.page_content for doc in docs)

def answer_directly(query):
    response = generator(query)
    return response[0]["generated_text"]

def clarify_question():
    return "Can you please clarify your question?"
```

```python
def agent_decide(query):
    keywords = ["library", "attendance", "exam", "hostel", "timing", "rule"]

    if any(word in query.lower() for word in keywords):
        return "PDF"
    elif len(query.split()) < 3:
        return "CLARIFY"
    else:
        return "DIRECT"
```

```python
def agent_answer(query):
    decision = agent_decide(query)

    if decision == "PDF":
        context = search_pdf(query)
        prompt = f"Answer using the context:\n{context}\n\nQuestion: {query}"
        return generator(prompt)[0]["generated_text"]

    elif decision == "DIRECT":
        return answer_directly(query)

    else:
        return clarify_question()
```

```python
print("Agentic Campus Assistant (type 'exit' to quit)")

while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        break

    response = agent_answer(user_input)
    print("AI:", response)
```
